```python
import nltk
import jieba
import pandas as pd
from tqdm import tqdm_notebook
from nltk.util import ngrams
from collections import Counter
from nltk.corpus import stopwords
from gensim.models.word2vec import Word2Vec
jieba.set_dictionary('/Users/andy/Documents/python/校外課程&自我學習/jieba//dict.txt.big.txt')
stop=stopwords.words('/Users/andy/Documents/python/校外課程&自我學習/jieba/stop_words2.txt')
```

```python
df_car = pd.read_excel('PTT汽車.xlsx')
df_cchat = pd.read_excel('PTT西洽.xlsx')
df_baseball = pd.read_excel('PTT棒球版.xlsx')
df_gossiping = pd.read_excel('PTT八卦.xlsx')
```

## Car

```python
text = []
for index in tqdm_notebook(range(len(df_car))):
    words = jieba.cut(str(df_car['回文'][index]))
    for word in words:
        if word != ' ':
            text.append(word)
```

```python
len(text)
```

```python
bigram = ngrams(text,2)
trgram = ngrams(text,3)
bi = Counter(bigram).most_common()
tr = Counter(trgram).most_common()
```

```python
new_bi = []
new_tr = []
stop = ['，','、','：','(',')','。',':','【','】','/','-','_',',','“','___','「','」','+','__','？','.','?','!','~','>','=','...','…']
for i in range(len(bi)):
    number = 0
    for s in stop:
        if s in bi[i][0]:
            number += 1
    if number == 0:
        new_bi.append(bi[i])

for i in range(len(tr)):
    number = 0
    for s in stop:
        if s in tr[i][0]:
            number += 1
    if number == 0:
        new_tr.append(tr[i])
```

```python
len(new_bi)
```

```python
new_bi[0:200]
```

```python
file = open("bigram_car.txt","w")
for i in new_bi:
    file.write(str(i)+"\n")
file.close()
```

```python
file = open("trigram_car.txt","w")
for i in new_tr:
    file.write(str(i)+"\n")
file.close()
```
